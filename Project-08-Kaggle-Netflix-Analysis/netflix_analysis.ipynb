# 📥 Step 1: Import NumPy

NumPy provides powerful array operations and numerical analysis functions.

Although this dataset contains many text columns, NumPy will still be useful for numerical calculations and statistical analysis.

```python
import numpy as np
```

# 📂 Step 2: Load the Dataset

The Netflix dataset is stored as a CSV file.

We use **np.genfromtxt()** to load the data.

Since the dataset contains text values, we load it as strings.

```python
netflix_data = np.genfromtxt(
    "dataset/netflix_titles.csv",
    delimiter=",",
    dtype=str,
    encoding="utf-8",
    skip_header=1
)

print(netflix_data)
```

Before starting the analysis, let's inspect the dataset.

We will determine:

- Shape

- Total Elements
- Data Type

```python
print("Shape :", netflix_data.shape)

print("Dimensions :", netflix_data.ndim)

print("Total Elements :", netflix_data.size)

print("Data Type :", netflix_data.dtype)
```

# 📊 Step 8: Movies vs TV Shows Percentage

Absolute numbers are useful, but percentages provide a better understanding of the content distribution.

## Formula

Percentage = (Category Count / Total Content) × 100

## Expected Outcome

- Movie Percentage
- TV Show Percentage

```python
total = movies + tv_shows

movie_percentage = (movies / total) * 100

tv_percentage = (tv_shows / total) * 100

print("=" * 50)
print("CONTENT PERCENTAGE")
print("=" * 50)

print(f"Movies   : {movie_percentage:.2f}%")

print(f"TV Shows : {tv_percentage:.2f}%")
```

# 📅 Step 9: Release Year Analysis

Release Year indicates when a movie or TV show was originally released.

Business Questions:

- What is the oldest title?
- What is the newest title?
- What is the average release year?

```python
release_year = netflix_data[:, 7].astype(int)

print("=" * 50)
print("RELEASE YEAR ANALYSIS")
print("=" * 50)

print(f"Oldest Release Year : {np.min(release_year)}")

print(f"Latest Release Year : {np.max(release_year)}")

print(f"Average Release Year : {np.mean(release_year):.2f}")

print(f"Median Release Year : {np.median(release_year):.0f}")

print(f"Standard Deviation : {np.std(release_year):.2f}")
```

# 🏆 Step 10: Oldest and Latest Titles

Finding the oldest and newest content helps us understand the historical range of Netflix's library.

## Business Question

Which titles are the oldest and newest in the dataset?

```python
title = netflix_data[:, 2]

oldest_index = np.argmin(release_year)

latest_index = np.argmax(release_year)

print("=" * 60)
print("OLDEST TITLE")
print("=" * 60)

print(title[oldest_index])

print()

print("=" * 60)
print("LATEST TITLE")
print("=" * 60)

print(title[latest_index])
```

# ⭐ Step 11: Rating Analysis

Ratings indicate the target audience for each title.

Examples include:

- TV-MA
- TV-14
- PG
- PG-13
- R

Business Question:

Which rating appears most frequently on Netflix?

```python
ratings = netflix_data[:, 8]

unique_ratings, counts = np.unique(ratings, return_counts=True)

print("=" * 50)
print("NETFLIX RATINGS")
print("=" * 50)

for rating, count in zip(unique_ratings, counts):
    print(f"{rating:<10} : {count}")
```

# 🌍 Step 12: Country Analysis

Netflix content is produced by many different countries.

Business Questions

- How many unique countries are represented?
- Which country appears most frequently?

```python
countries = netflix_data[:, 5]

countries = countries[countries != ""]

unique_countries, country_counts = np.unique(countries, return_counts=True)

most_country = unique_countries[np.argmax(country_counts)]

print("=" * 50)
print("COUNTRY ANALYSIS")
print("=" * 50)

print(f"Unique Countries : {len(unique_countries)}")

print(f"Most Frequent Country : {most_country}")
```

# 📈 Step 13: Top 10 Release Years

This analysis identifies the years in which Netflix has the highest number of available titles.

Understanding release trends helps reveal how content production has evolved over time.

```python
years, counts = np.unique(release_year, return_counts=True)

sorted_index = np.argsort(counts)[::-1]

print("=" * 60)
print("TOP 10 RELEASE YEARS")
print("=" * 60)

for i in sorted_index[:10]:
    print(f"{years[i]} : {counts[i]} titles")
```

# ⏱️ Step 14: Duration Analysis

The duration column represents:

- **Movies** → Duration in minutes (e.g., 90 min)
- **TV Shows** → Number of seasons (e.g., 3 Seasons)

## Business Questions

- What is the average movie duration?
- What is the longest movie?
- What is the shortest movie?

For this analysis, we consider only Movies.

```python
duration = netflix_data[:, 9]

movie_duration = []

for i in range(len(content_type)):
    if content_type[i] == "Movie":
        try:
            minutes = int(duration[i].split()[0])
            movie_duration.append(minutes)
        except:
            pass

movie_duration = np.array(movie_duration)

print("="*50)
print("MOVIE DURATION ANALYSIS")
print("="*50)

print("Average Duration :", np.mean(movie_duration), "minutes")
print("Longest Movie :", np.max(movie_duration), "minutes")
print("Shortest Movie :", np.min(movie_duration), "minutes")
print("Median Duration :", np.median(movie_duration), "minutes")
```

# 📊 Step 15: Duration Statistics

Let's understand the spread of movie durations using statistical measures.

```python
print("="*50)
print("MOVIE DURATION STATISTICS")
print("="*50)

print("Standard Deviation :", np.std(movie_duration))

print("25th Percentile :", np.percentile(movie_duration,25))

print("50th Percentile :", np.percentile(movie_duration,50))

print("75th Percentile :", np.percentile(movie_duration,75))
```

# 🔝 Step 16: Top 10 Longest Movies

Finding the longest movies available on Netflix.

```python
sorted_duration = np.sort(movie_duration)

print("="*50)
print("TOP 10 LONGEST MOVIES")
print("="*50)

print(sorted_duration[-10:])
```
