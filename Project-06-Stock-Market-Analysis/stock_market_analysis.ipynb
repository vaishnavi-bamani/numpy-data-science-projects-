# 📥 Step 1: Import NumPy Library

Before analyzing stock market data, we first import the NumPy library.

NumPy provides efficient multidimensional arrays and high-performance mathematical operations, making it ideal for financial data analysis.

## NumPy Concept Used

- Importing Libraries

## Expected Outcome

The NumPy library will be available using the alias **np** throughout the notebook.

```python
import numpy as np
```

# 📂 Step 2: Load Stock Market Dataset

The stock market dataset is stored in a CSV file.

Using **np.genfromtxt()**, we load the numerical data into a NumPy array for further analysis.

## NumPy Function Used

- np.genfromtxt()

## Expected Outcome

The CSV file will be converted into a two-dimensional NumPy array.

```python
stock_data = np.genfromtxt(
    "dataset/stock_market.csv",
    delimiter=",",
    skip_header=1
)

print(stock_data)
```

# 🔍 Step 3: Explore the Dataset

Before beginning analysis, it is important to inspect the dataset.

We will determine:

- Data Type
- Shape
- Number of Dimensions

This ensures the dataset has been loaded correctly.

```python
print(stock_data)

print("Data Type :", stock_data.dtype)

print("Shape :", stock_data.shape)

print("Dimensions :", stock_data.ndim)
```

# 📊 Step 4: Extract Individual Columns

The stock market dataset contains six columns.

We will separate each column into an independent NumPy array.

## Columns

- Day
- Opening Price
- Highest Price
- Lowest Price
- Closing Price
- Trading Volume

This makes financial analysis easier and more organized.

```python
day = stock_data[:,0]

open_price = stock_data[:,1]

high_price = stock_data[:,2]

low_price = stock_data[:,3]

close_price = stock_data[:,4]

volume = stock_data[:,5]
```

# 👀 Step 5: Display Extracted Data

Display every extracted column to verify that the data has been correctly separated.

This validation step ensures that each column has been extracted successfully before performing calculations.

```python
print("========== DAYS ==========\n")
print(day)

print("\n========== OPEN PRICE ==========\n")
print(open_price)

print("\n========== HIGH PRICE ==========\n")
print(high_price)

print("\n========== LOW PRICE ==========\n")
print(low_price)

print("\n========== CLOSE PRICE ==========\n")
print(close_price)

print("\n========== VOLUME ==========\n")
print(volume)
```

# 💹 Step 6: Closing Price Analysis

The closing price is one of the most important indicators in stock market analysis because it represents the final trading price of a stock for the day.

We will calculate:

- Highest Closing Price
- Lowest Closing Price
- Average Closing Price
- Median Closing Price
- Standard Deviation
- Highest Closing Day
- Lowest Closing Day

```python
highest_close = np.max(close_price)
lowest_close = np.min(close_price)

average_close = np.mean(close_price)
median_close = np.median(close_price)

std_close = np.std(close_price)

highest_close_day = day[np.argmax(close_price)]
lowest_close_day = day[np.argmin(close_price)]

print("========== CLOSING PRICE ANALYSIS ==========\n")

print(f"Highest Closing Price : ₹{highest_close:.2f}")
print(f"Lowest Closing Price  : ₹{lowest_close:.2f}")
print(f"Average Closing Price : ₹{average_close:.2f}")
print(f"Median Closing Price  : ₹{median_close:.2f}")
print(f"Standard Deviation    : {std_close:.2f}")

print()

print(f"Highest Closing Day : Day {int(highest_close_day)}")
print(f"Lowest Closing Day  : Day {int(lowest_close_day)}")
```

# 📈 Step 7: Daily Price Change

The daily price change measures how much the stock price changed during a trading day.

Formula:

Price Change = Closing Price − Opening Price

A positive value indicates the stock gained value during the day, while a negative value indicates a loss.

```python
price_change = close_price - open_price

print("========== DAILY PRICE CHANGE ==========\n")

for d, change in zip(day, price_change):
    print(f"Day {int(d)} : {change:.2f}")
```

# 📊 Step 8: Daily Percentage Return

Percentage Return provides a better measure of daily performance because it compares the price change relative to the opening price.

Formula:

Percentage Return = (Price Change / Opening Price) × 100

```python
percentage_return = (price_change / open_price) * 100

print("========== DAILY PERCENTAGE RETURN ==========\n")

for d, ret in zip(day, percentage_return):
    print(f"Day {int(d)} : {ret:.2f}%")
```

# 📦 Step 9: Trading Volume Analysis

Trading volume indicates the total number of shares traded during a day.

We will calculate:

- Highest Trading Volume
- Lowest Trading Volume
- Average Trading Volume

```python
print("========== TRADING VOLUME ANALYSIS ==========\n")

print(f"Highest Volume : {np.max(volume):,.0f}")

print(f"Lowest Volume  : {np.min(volume):,.0f}")

print(f"Average Volume : {np.mean(volume):,.2f}")
```

# 🐂 Step 10: Bullish and Bearish Days

A trading day is classified as:

- Bullish → Closing Price > Opening Price
- Bearish → Closing Price < Opening Price

This helps identify the market trend over the 30-day period.

```python
bullish_days = np.sum(close_price > open_price)

bearish_days = np.sum(close_price < open_price)

print("========== MARKET TREND ==========\n")

print(f"Bullish Days : {bullish_days}")

print(f"Bearish Days : {bearish_days}")
```

# 📉 Step 11: Price Volatility

Volatility measures how much stock prices fluctuate over time.

We calculate volatility using the standard deviation of closing prices.

Higher volatility indicates greater price fluctuations.

```python
volatility = np.std(close_price)

print("========== PRICE VOLATILITY ==========\n")

print(f"Stock Price Volatility : {volatility:.2f}")
```

# 🏆 Step 12: Top Five Closing Prices

Sorting closing prices helps identify the highest prices reached during the analysis period.

We display the top five closing prices.

top_five = np.sort(close_price)[-5:]

print("========== TOP FIVE CLOSING PRICES ==========\n")

print(top_five)

# 📊 Step 13: Closing Price Percentiles

Percentiles help understand the distribution of closing prices.

We calculate:

- 25th Percentile
- 50th Percentile (Median)
- 75th Percentile

```python
print("========== PRICE PERCENTILES ==========\n")

print(f"25th Percentile : {np.percentile(close_price,25):.2f}")

print(f"50th Percentile : {np.percentile(close_price,50):.2f}")

print(f"75th Percentile : {np.percentile(close_price,75):.2f}")
```

# 📋 Step 14: Stock Market Dashboard

The dashboard summarizes all important statistics generated throughout the project.

```python
print("="*70)
print("           STOCK MARKET ANALYSIS DASHBOARD")
print("="*70)

print(f"Trading Days              : {len(day)}")

print(f"Highest Closing Price     : ₹{highest_close:.2f}")

print(f"Lowest Closing Price      : ₹{lowest_close:.2f}")

print(f"Average Closing Price     : ₹{average_close:.2f}")

print(f"Median Closing Price      : ₹{median_close:.2f}")

print(f"Price Volatility          : {volatility:.2f}")

print()

print(f"Highest Trading Volume    : {np.max(volume):,.0f}")

print(f"Lowest Trading Volume     : {np.min(volume):,.0f}")

print(f"Average Trading Volume    : {np.mean(volume):,.2f}")

print()

print(f"Bullish Days              : {bullish_days}")

print(f"Bearish Days              : {bearish_days}")

print()

print(f"Highest Closing Day       : Day {int(highest_close_day)}")

print(f"Lowest Closing Day        : Day {int(lowest_close_day)}")

print("="*70)
print("End of Report")
print("="*70)
```
