# 📥 Step 1: Import NumPy

NumPy is the core library used for image manipulation because an image is simply a multidimensional NumPy array.

We first import NumPy before loading the dataset.

```python
import numpy as np
```

# 📂 Step 2: Load Image Dataset

The grayscale image is stored as a CSV file.

We use **np.genfromtxt()** to load the image into a two-dimensional NumPy array.

```python
image = np.genfromtxt(
    "dataset/grayscale_image.csv",
    delimiter=","
)

print(image)
```

# 🔍 Step 3: Explore the Image

Before processing the image, inspect its basic properties.

We will determine:

- Data Type
- Shape
- Number of Dimensions
- Total Number of Pixels

```python
print("Image Shape :", image.shape)

print("Dimensions :", image.ndim)

print("Data Type :", image.dtype)

print("Total Pixels :", image.size)
```

# 🌈 Step 4: Display Pixel Matrix

Since the image is represented as a matrix, we display every pixel intensity.

Each value represents the brightness of one pixel.

```python
print(image)
```

# 📊 Step 5: Image Statistics

Image statistics help us understand the overall characteristics of the image.

We calculate:

- Brightest Pixel
- Darkest Pixel
- Average Brightness
- Median Brightness
- Standard Deviation

```python
print("========== IMAGE STATISTICS ==========\n")

print("Brightest Pixel :", np.max(image))

print("Darkest Pixel :", np.min(image))

print("Average Brightness :", np.mean(image))

print("Median Brightness :", np.median(image))

print("Standard Deviation :", np.std(image))
```

# ☀️ Step 6: Increase Brightness

Brightness adjustment is performed by adding a constant value to every pixel.

Formula:

New Pixel = Old Pixel + Brightness Value

Since pixel values cannot exceed **255**, we use **np.clip()**.

```python
bright_image = np.clip(image + 30, 0, 255)

print(bright_image)
```

# 🌙 Step 7: Decrease Brightness

To darken an image, subtract a constant value from every pixel.

Again, **np.clip()** ensures pixel values remain between **0** and **255**.

```python
dark_image = np.clip(image - 30, 0, 255)

print(dark_image)
```

# 🎨 Step 8: Contrast Enhancement

Contrast determines the difference between dark and bright pixels in an image.

Increasing contrast makes bright pixels brighter and dark pixels darker.

Formula:

New Pixel = Pixel × Contrast Factor

Since pixel values must remain between **0 and 255**, we use **np.clip()**.

```python
contrast_image = np.clip(image * 1.5, 0, 255)

print("========== CONTRAST ENHANCED IMAGE ==========\n")

print(contrast_image)
```

# 🎭 Step 9: Negative Image

A negative image is created by subtracting every pixel value from **255**.

Formula:

Negative Pixel = 255 − Original Pixel

This operation is widely used in medical imaging and computer vision.

```python
negative_image = 255 - image

print("========== NEGATIVE IMAGE ==========\n")

print(negative_image)
```

# ⚫ Step 10: Binary Thresholding

Thresholding converts a grayscale image into a binary image.

Rule:

- Pixel ≥ 100 → 255 (White)
- Pixel < 100 → 0 (Black)

This technique is commonly used for image segmentation and object detection.

```python
binary_image = np.where(image >= 100, 255, 0)

print("========== BINARY IMAGE ==========\n")

print(binary_image)
```

# ✂️ Step 11: Image Cropping

Cropping extracts a selected region from the image.

Here, we extract the center **4 × 4** portion of the image.

```python
cropped_image = image[2:6, 2:6]

print("========== CROPPED IMAGE ==========\n")

print(cropped_image)
```

# ↔️ Step 12: Horizontal Flip

Horizontal flipping mirrors the image from left to right.

NumPy Function:

- np.fliplr()

```python
horizontal_flip = np.fliplr(image)

print("========== HORIZONTAL FLIP ==========\n")

print(horizontal_flip)
```

# ↕️ Step 13: Vertical Flip

Vertical flipping mirrors the image from top to bottom.

NumPy Function:

- np.flipud()

```python
vertical_flip = np.flipud(image)

print("========== VERTICAL FLIP ==========\n")

print(vertical_flip)
```

# 🔄 Step 14: Rotate Image by 90°

Rotation changes the orientation of an image.

NumPy Function:

- np.rot90()

The image is rotated 90° counter-clockwise.

```python
rotated_image = np.rot90(image)

print("========== ROTATED IMAGE ==========\n")

print(rotated_image)
```

# 📏 Step 15: Normalize Pixel Values

Normalization converts pixel values from the range **0–255** to **0–1**.

Formula:

Normalized Pixel = Pixel / 255

Normalization is an important preprocessing step in Deep Learning and Computer Vision.

```python
normalized_image = image / 255

print("========== NORMALIZED IMAGE ==========\n")

print(normalized_image)
```

# 📋 Step 16: Image Processing Dashboard

Summarize all important statistics and processing results generated throughout the project.

```python
print("=" * 70)
print("          IMAGE PROCESSING DASHBOARD")
print("=" * 70)

print(f"Image Shape               : {image.shape}")

print(f"Total Pixels              : {image.size}")

print(f"Brightest Pixel           : {np.max(image)}")

print(f"Darkest Pixel             : {np.min(image)}")

print(f"Average Brightness        : {np.mean(image):.2f}")

print(f"Median Brightness         : {np.median(image):.2f}")

print(f"Standard Deviation        : {np.std(image):.2f}")

print()

print(f"Maximum Normalized Value  : {np.max(normalized_image):.2f}")

print(f"Minimum Normalized Value  : {np.min(normalized_image):.2f}")

print()

print(f"Cropped Image Shape       : {cropped_image.shape}")

print("=" * 70)
print("End of Report")
print("=" * 70)
```
