```python
import numpy as np
```

# 📂 Load Dataset

Read the CSV file using **NumPy** and convert it into a multidimensional array for analysis.

```python
sales_data = np.genfromtxt(
    "dataset/ECommerce.csv",
    delimiter=",",
    skip_header=1
)

print(sales_data)
```

# 🔍 Explore Dataset

Inspect the dataset before analysis.

We will determine:

- Data Type
- Shape
- Number of Dimensions

```python
print("Data Type :", sales_data.dtype)

print("Shape :", sales_data.shape)

print("Dimensions :", sales_data.ndim)
```

# 📊 Extract Individual Columns

Separate each column into individual NumPy arrays.

```python
order_id = sales_data[:,0]

product_id = sales_data[:,1]

category = sales_data[:,2]

quantity = sales_data[:,3]

price = sales_data[:,4]

discount = sales_data[:,5]
```

# 👀 Display Extracted Data

Verify each extracted array before analysis.

```python
print(order_id)

print(product_id)

print(category)

print(quantity)

print(price)

print(discount)
```

# 💰 Revenue Calculation

Gross Revenue is calculated as:

Revenue = Quantity × Price

```python
gross_revenue = quantity * price

print(gross_revenue)
```

# 🏷 Discount Analysis

Calculate the discount amount for every order.

```python
discount_amount = gross_revenue * (discount/100)

print(discount_amount)
```

# 💵 Net Revenue

Net Revenue = Gross Revenue − Discount

```python
net_revenue = gross_revenue - discount_amount

print(net_revenue)
```

# 📈 Revenue Statistics

Calculate:

- Total Revenue
- Highest Revenue
- Lowest Revenue
- Average Revenue
- Median Revenue
- Standard Deviation

```python
print("Total Revenue :", np.sum(net_revenue))

print("Highest Revenue :", np.max(net_revenue))

print("Lowest Revenue :", np.min(net_revenue))

print("Average Revenue :", np.mean(net_revenue))

print("Median Revenue :", np.median(net_revenue))

print("Standard Deviation :", np.std(net_revenue))
```

# 🏆 Highest Revenue Order

Identify the order that generated the highest revenue.

```python
highest_order = order_id[np.argmax(net_revenue)]

lowest_order = order_id[np.argmin(net_revenue)]

print("Highest Revenue Order :", highest_order)

print("Lowest Revenue Order :", lowest_order)
```

# 🛍 Category-wise Analysis

Categories

1 → Electronics

2 → Clothing

3 → Grocery

```python
electronics = np.sum(category==1)

clothing = np.sum(category==2)

grocery = np.sum(category==3)

print("Electronics Orders :", electronics)

print("Clothing Orders :", clothing)

print("Grocery Orders :", grocery)
```

# ⭐ High Value Orders

Orders generating more than ₹5000 are classified as High Value Orders.

```python
high_value_orders = sales_data[net_revenue>5000]

print(high_value_orders)
```

# 📊 Revenue Classification

Classify revenue into:

- Low
- Medium
- High

using **np.where()**.

```python
revenue_category = np.where(
    net_revenue<3000,
    "Low",
    np.where(
        net_revenue<=7000,
        "Medium",
        "High"
    )
)

for order,revenue,status in zip(order_id,net_revenue,revenue_category):

    print(f"Order {int(order)} : ₹{revenue:.2f} --> {status}")
```

# 🥇 Top Five Revenue Orders

Display the highest five revenues.

```python
top_five = np.sort(net_revenue)[-5:]

print(top_five)
```

# 📈 Revenue Distribution

Calculate important percentiles.

```python
print("25th Percentile :", np.percentile(net_revenue,25))

print("50th Percentile :", np.percentile(net_revenue,50))

print("75th Percentile :", np.percentile(net_revenue,75))
```

# 📋 Final Sales Dashboard

Summarize the complete project.

```python
print("="*65)
print("        E-COMMERCE SALES ANALYSIS DASHBOARD")
print("="*65)

print(f"Total Orders              : {len(order_id)}")

print(f"Total Revenue             : ₹{np.sum(net_revenue):,.2f}")

print(f"Highest Revenue           : ₹{np.max(net_revenue):,.2f}")

print(f"Lowest Revenue            : ₹{np.min(net_revenue):,.2f}")

print(f"Average Revenue           : ₹{np.mean(net_revenue):,.2f}")

print(f"Median Revenue            : ₹{np.median(net_revenue):,.2f}")

print(f"Revenue Std Deviation     : {np.std(net_revenue):,.2f}")

print()

print(f"Electronics Orders        : {electronics}")

print(f"Clothing Orders           : {clothing}")

print(f"Grocery Orders            : {grocery}")

print()

print(f"High Value Orders         : {len(high_value_orders)}")

print()

print(f"Highest Revenue Order ID  : {int(highest_order)}")

print(f"Lowest Revenue Order ID   : {int(lowest_order)}")

print("="*65)
print("End of Report")
print("="*65)
```
