# 📥 Step 1: Import NumPy Library

Before performing any data analysis, we first import the NumPy library.

NumPy provides powerful tools for handling numerical data efficiently using multidimensional arrays and vectorized operations.

## NumPy Concept Used

- Importing Libraries

## Expected Outcome

The NumPy library will be available throughout the notebook using the alias **np**.

```python
import numpy as np
```

# 📂 Step 2: Load Employee Dataset

The employee salary dataset is stored as a CSV file.

Using **np.genfromtxt()**, we load the numerical data into a NumPy array.

## NumPy Function Used

- np.genfromtxt()

## Expected Outcome

The CSV file will be converted into a two-dimensional NumPy array for further analysis.

```python
employee_data = np.genfromtxt(
    "dataset/employee_salary.csv",
    delimiter=",",
    skip_header=1
)

print(employee_data)
```

# 🔍 Step 3: Explore the Dataset

Before analysis, it is important to inspect the dataset.

We will determine:

- Data Type
- Shape
- Number of Dimensions

This helps verify that the dataset has been loaded correctly.

```python
print(employee_data)

print("Data Type:", employee_data.dtype)
print("Shape:", employee_data.shape)
print("Dimensions:", employee_data.ndim)
```

# 📊 Step 4: Extract Individual Columns

The dataset consists of five columns.

We will separate each column into an independent NumPy array.

## Columns

- Employee ID
- Age
- Experience
- Department
- Salary

This makes further analysis much easier.

```python
employee_id = employee_data[:,0]

age = employee_data[:,1]

experience = employee_data[:,2]

department = employee_data[:,3]

salary = employee_data[:,4]
```

# 👀 Step 5: Display Extracted Data

Display every extracted column to verify that the data has been correctly separated.

This is considered a good data validation practice before beginning any statistical analysis.

```python
print("========== EMPLOYEE IDs ==========\n")
print(employee_id)

print("\n========== AGE ==========\n")
print(age)

print("\n========== EXPERIENCE ==========\n")
print(experience)

print("\n========== DEPARTMENT ==========\n")
print(department)

print("\n========== SALARY ==========\n")
print(salary)
```

# 💰 Step 6: Salary Analysis

Salary is the primary metric in this project.

We will calculate:

- Highest Salary
- Lowest Salary
- Average Salary
- Median Salary
- Standard Deviation
- Highest Paid Employee
- Lowest Paid Employee

```python
highest_salary = np.max(salary)
lowest_salary = np.min(salary)

average_salary = np.mean(salary)
median_salary = np.median(salary)

std_salary = np.std(salary)

highest_paid_employee = employee_id[np.argmax(salary)]
lowest_paid_employee = employee_id[np.argmin(salary)]

print("========== SALARY ANALYSIS ==========\n")

print(f"Highest Salary : ₹{highest_salary:,.0f}")
print(f"Lowest Salary  : ₹{lowest_salary:,.0f}")
print(f"Average Salary : ₹{average_salary:,.2f}")
print(f"Median Salary  : ₹{median_salary:,.2f}")
print(f"Standard Deviation : {std_salary:.2f}")

print()

print(f"Highest Paid Employee ID : {int(highest_paid_employee)}")
print(f"Lowest Paid Employee ID  : {int(lowest_paid_employee)}")
```

# 👨‍💼 Step 7: Age Analysis

Analyze employee ages by calculating:

- Maximum Age
- Minimum Age
- Average Age

These statistics provide insights into workforce demographics.

```python
print("========== AGE ANALYSIS ==========\n")

print(f"Maximum Age : {np.max(age):.0f}")

print(f"Minimum Age : {np.min(age):.0f}")

print(f"Average Age : {np.mean(age):.2f}")
```

# 📈 Step 8: Experience Analysis

Employee experience is another important HR metric.

We will calculate:

- Maximum Experience
- Minimum Experience
- Average Experience

```python
print("========== EXPERIENCE ANALYSIS ==========\n")

print(f"Maximum Experience : {np.max(experience):.0f} Years")

print(f"Minimum Experience : {np.min(experience):.0f} Years")

print(f"Average Experience : {np.mean(experience):.2f} Years")
```

# 🏢 Step 9: Department Analysis

Employees belong to different departments.

Department Codes

1 → IT

2 → HR

3 → Finance

We calculate the number of employees in each department.

```python
print("========== DEPARTMENT ANALYSIS ==========\n")

print("IT Employees :", np.sum(department==1))

print("HR Employees :", np.sum(department==2))

print("Finance Employees :", np.sum(department==3))
```

# 💵 Step 10: Salary Classification

Using **np.where()**, salaries are classified into:

- Low Salary
- Medium Salary
- High Salary

This demonstrates conditional selection using NumPy.

```python
salary_category = np.where(
    salary<50000,
    "Low",
    np.where(
        salary<=80000,
        "Medium",
        "High"
    )
)

for emp,sal,cat in zip(employee_id,salary,salary_category):
    print(f"Employee {int(emp)} : ₹{sal:,.0f} --> {cat}")
```

# ⭐ Step 11: Employees Earning Above Average Salary

Boolean masking is used to identify employees whose salaries are above the company average.

```python
above_average = employee_data[salary>average_salary]

print(above_average)
```

# 🥇 Step 12: Top Five Salaries

Sort the salary array and display the five highest salaries.

```python
top_five = np.sort(salary)[-5:]

print("Top Five Salaries")

print(top_five)
```

# 📋 Step 13: Employee Salary Dashboard

The dashboard summarizes all major statistics generated throughout the project.

```python
print("="*65)
print("         EMPLOYEE SALARY ANALYSIS DASHBOARD")
print("="*65)

print(f"Total Employees          : {len(employee_id)}")

print(f"Highest Salary           : ₹{highest_salary:,.0f}")

print(f"Lowest Salary            : ₹{lowest_salary:,.0f}")

print(f"Average Salary           : ₹{average_salary:,.2f}")

print(f"Median Salary            : ₹{median_salary:,.2f}")

print(f"Salary Std Deviation     : {std_salary:.2f}")

print()

print(f"Maximum Age              : {np.max(age):.0f}")

print(f"Minimum Age              : {np.min(age):.0f}")

print(f"Average Age              : {np.mean(age):.2f}")

print()

print(f"Maximum Experience       : {np.max(experience):.0f}")

print(f"Average Experience       : {np.mean(experience):.2f}")

print()

print(f"IT Employees             : {np.sum(department==1)}")

print(f"HR Employees             : {np.sum(department==2)}")

print(f"Finance Employees        : {np.sum(department==3)}")

print()

print(f"Employees Above Average Salary : {len(above_average)}")

print("="*65)
print("End of Report")
print("="*65)
```
