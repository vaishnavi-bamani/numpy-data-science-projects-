# 🎓 Student Performance Analysis using NumPy

- Perform subject-wise analysis
- Analyze pass and fail percentages

- Generate a complete class performance report

# 📚 Step 1: Import NumPy Library

NumPy is the fundamental Python library for numerical computing.

It provides:

- Fast array operations
- Mathematical functions
- Statistical calculations
- Efficient data analysis

```python
# Import NumPy library
import numpy as np
```

# 📂 Step 2: Load the Dataset

The dataset contains marks obtained by 30 students in three subjects:

- Mathematics
- Science
- English

The dataset is stored in CSV format.

We use **np.genfromtxt()** to load the dataset into a NumPy array.

```python
# Load the student marks dataset

student_data = np.genfromtxt(
    "dataset/student_marks.csv",
    delimiter=",",
    skip_header=1
)
```

# 🔍 Step 3: Explore the Dataset

Before starting any analysis, it is important to understand the dataset.

We will check:

- Complete dataset
- Data type
- Shape
- Dimensions

```python
# Display the complete dataset

print("========== STUDENT DATASET ==========\n")
print(student_data)

print("\n========== DATASET INFORMATION ==========")

print("Type :", type(student_data))
print("Shape :", student_data.shape)
print("Data Type :", student_data.dtype)
```

# 📖 Step 4: Extract Individual Columns

The dataset contains four columns.

| Column | Description |
|---------|-------------|
| Student | Student ID |
| Math | Mathematics Marks |
| Science | Science Marks |
| English | English Marks |

For easier analysis, each column is extracted into a separate NumPy array.

```python
# Extract each column

students = student_data[:, 0]

math = student_data[:, 1]

science = student_data[:, 2]

english = student_data[:, 3]
```

# 👀 Step 5: Display Extracted Data

Now we display the extracted arrays to verify that each column has been loaded correctly.

```python
print("Student IDs")
print(students)

print("\nMath Marks")
print(math)

print("\nScience Marks")
print(science)

print("\nEnglish Marks")
print(english)
```

# 📊 Step 6: Calculate Total Marks

Each student's total marks are calculated by adding their marks in Mathematics, Science, and English.

**Formula**

Total Marks = Math + Science + English

```python
# Calculate total marks of each student

total_marks = math + science + english

print("========== TOTAL MARKS ==========\n")

for student, total in zip(students, total_marks):
    print(f"Student {int(student):2d} : {total:.0f}")
```

# 📈 Step 7: Calculate Average Marks

The average marks represent the overall academic performance of each student.

**Formula**

Average Marks = Total Marks ÷ Number of Subjects

```python
# Calculate average marks

average_marks = total_marks / 3

print("========== AVERAGE MARKS ==========\n")

for student, average in zip(students, average_marks):
    print(f"Student {int(student):2d} : {average:.2f}")
```

# 🏆 Step 8: Find the Highest and Lowest Scorer

NumPy provides efficient functions to identify the highest and lowest values in an array.

Functions used:

- np.max()
- np.min()
- np.argmax()
- np.argmin()

```python
# Highest scorer

highest_marks = np.max(total_marks)
highest_index = np.argmax(total_marks)
highest_student = students[highest_index]

# Lowest scorer

lowest_marks = np.min(total_marks)
lowest_index = np.argmin(total_marks)
lowest_student = students[lowest_index]

print("========== HIGHEST & LOWEST SCORER ==========\n")

print(f"Highest Scorer : Student {int(highest_student)}")
print(f"Total Marks    : {highest_marks:.0f}")

print()

print(f"Lowest Scorer  : Student {int(lowest_student)}")
print(f"Total Marks    : {lowest_marks:.0f}")
```

# ✅ Step 9: Pass and Fail Analysis

A student is considered **Pass** if the average marks are **40 or above**.

Otherwise, the student is considered **Fail**.

Boolean masking is used to efficiently separate pass and fail students.

```python
# Define passing marks

passing_marks = 40

# Create boolean masks

pass_mask = average_marks >= passing_marks
fail_mask = average_marks < passing_marks

# Count students

pass_students = np.sum(pass_mask)
fail_students = np.sum(fail_mask)

print("========== PASS / FAIL ANALYSIS ==========\n")

print(f"Passed Students : {pass_students}")
print(f"Failed Students : {fail_students}")
```

# 📋 Step 10: Display Pass and Fail Students

Using boolean indexing, we can directly retrieve the student IDs of those who passed and failed.

```python
# 📚 Step 11: Subject-wise Average Marks Analysis

## Business Problem

Teachers want to evaluate the overall class performance in each subject.

### NumPy Function Used

- `np.mean()`

### Expected Outcome

Calculate the average marks in Mathematics, Science, and English to understand subject-wise performance.
```

```python
# Calculate subject-wise average marks

math_average = np.mean(math)
science_average = np.mean(science)
english_average = np.mean(english)

print("========== SUBJECT-WISE AVERAGE MARKS ==========\n")

print(f"Math Average    : {math_average:.2f}")
print(f"Science Average : {science_average:.2f}")
print(f"English Average : {english_average:.2f}")
```

### 📈 Business Insight

Subject-wise averages help teachers understand overall class performance. A higher average indicates that students performed well in that subject, while a lower average may suggest that additional attention is needed.

```python
# Highest scorer in each subject

math_topper = students[np.argmax(math)]
science_topper = students[np.argmax(science)]
english_topper = students[np.argmax(english)]

print("========== SUBJECT TOPPERS ==========\n")

print(f"Math    : Student {int(math_topper)} ({np.max(math):.0f})")
print(f"Science : Student {int(science_topper)} ({np.max(science):.0f})")
print(f"English : Student {int(english_topper)} ({np.max(english):.0f})")
```

### 📈 Business Insight

Recognizing the highest scorer in each subject helps identify subject-specific strengths and motivates students to perform better.

```python
# 📉 Step 13: Lowest Scorer in Each Subject

## Business Problem

Teachers also need to identify students who require additional academic support.

### NumPy Functions Used

- `np.min()`
- `np.argmin()`

### Expected Outcome

Find the lowest marks and corresponding student in each subject.
```

```python
# Lowest scorer in each subject

math_lowest = students[np.argmin(math)]
science_lowest = students[np.argmin(science)]
english_lowest = students[np.argmin(english)]

print("========== LOWEST SCORERS ==========\n")

print(f"Math    : Student {int(math_lowest)} ({np.min(math):.0f})")
print(f"Science : Student {int(science_lowest)} ({np.min(science):.0f})")
print(f"English : Student {int(english_lowest)} ({np.min(english):.0f})")
```

### 📈 Business Insight

Identifying the lowest scorer in each subject helps teachers provide targeted guidance and improve students' learning outcomes.

# ⭐ Step 14: Easiest and Toughest Subject

## Business Problem

The school wants to determine which subject students performed best in and which subject they found most challenging.

### NumPy Functions Used

- `np.mean()`
- `np.argmax()`
- `np.argmin()`

### Expected Outcome

Compare subject averages to identify the easiest and toughest subject.

```python
# Store subject averages

subject_names = np.array(["Math", "Science", "English"])
subject_averages = np.array([
    math_average,
    science_average,
    english_average
])

# Find easiest and toughest subjects

easiest_subject = subject_names[np.argmax(subject_averages)]
toughest_subject = subject_names[np.argmin(subject_averages)]

print("========== SUBJECT ANALYSIS ==========\n")

print(f"Easiest Subject : {easiest_subject}")
print(f"Toughest Subject: {toughest_subject}")
```

### 📈 Business Insight

The easiest subject has the highest class average, indicating that students performed well overall. The toughest subject has the lowest average, suggesting that students may need additional support or improvements in teaching strategies.

# 🌟 Step 15: Students Scoring Above Class Average

## Business Problem

The school wants to identify students who performed better than the overall class average.

### NumPy Concepts Used

- `np.mean()`
- Boolean Masking

### Expected Outcome

Display the students whose average marks are higher than the class average.

```python
# Calculate class average

class_average = np.mean(average_marks)

# Create a boolean mask
above_average_mask = average_marks > class_average

# Extract students
above_average_students = students[above_average_mask]
above_average_marks = average_marks[above_average_mask]

print("========== STUDENTS ABOVE CLASS AVERAGE ==========\n")

print(f"Class Average : {class_average:.2f}\n")

for student, avg in zip(above_average_students, above_average_marks):
    print(f"Student {int(student):2d} : {avg:.2f}")
```

### 📈 Business Insight

Students scoring above the class average are performing better than most of the class and may be recognized for their academic excellence.

# 🥇 Step 16: Top 5 Performing Students

## Business Problem

Schools often recognize the top-performing students based on their overall academic performance.

### NumPy Functions Used

- `np.argsort()`

### Expected Outcome

Display the Top 5 students based on total marks.

```python
# Get indices of Top 5 students

top5_indices = np.argsort(total_marks)[-5:][::-1]

# Extract data

top5_students = students[top5_indices]
top5_totals = total_marks[top5_indices]
top5_averages = average_marks[top5_indices]

print("========== TOP 5 STUDENTS ==========\n")

for rank, (student, total, average) in enumerate(
    zip(top5_students, top5_totals, top5_averages),
    start=1
):
    print(
        f"Rank {rank} | "
        f"Student {int(student):2d} | "
        f"Total = {total:.0f} | "
        f"Average = {average:.2f}"
    )
```

### 📈 Business Insight

Ranking students based on total marks helps schools identify outstanding performers for awards, scholarships, and academic recognition.

# 📋 Step 17: Student Performance Dashboard

The final dashboard summarizes all important statistics generated during the analysis.

This provides a quick overview of the overall class performance.

```python
print("=" * 60)
print("        STUDENT PERFORMANCE REPORT")
print("=" * 60)

print(f"Total Students                 : {len(students)}")

print(f"Class Average                  : {class_average:.2f}")

print(f"Highest Scorer                 : Student {int(highest_student)}")
print(f"Highest Marks                  : {highest_marks:.0f}")

print(f"Lowest Scorer                  : Student {int(lowest_student)}")
print(f"Lowest Marks                   : {lowest_marks:.0f}")

print(f"Passed Students                : {pass_students}")
print(f"Failed Students                : {fail_students}")

print(f"Math Average                   : {math_average:.2f}")
print(f"Science Average                : {science_average:.2f}")
print(f"English Average                : {english_average:.2f}")

print(f"Easiest Subject                : {easiest_subject}")
print(f"Toughest Subject               : {toughest_subject}")

print("\nTop Performing Student")

print(f"Student ID                     : {int(top5_students[0])}")
print(f"Total Marks                    : {top5_totals[0]:.0f}")
print(f"Average Marks                  : {top5_averages[0]:.2f}")

print("=" * 60)
print("End of Report")
print("=" * 60)
```
