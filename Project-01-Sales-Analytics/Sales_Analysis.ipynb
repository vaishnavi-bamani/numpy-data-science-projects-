# 📊 Sales Data Analysis using NumPy

## Project Overview

This project analyzes one month of retail sales data using **NumPy**. The goal is to extract meaningful business insights from raw sales data using vectorized operations without traditional Python loops.

### 🎯 Objectives

- Calculate total monthly revenue
- Find average daily sales
- Identify highest and lowest sales days
- Analyze sales growth trends
- Compare weekend and weekday sales
- Evaluate sales target achievement
- Generate a business summary dashboard

### 🛠 Technologies Used

- Python
- NumPy
- Jupyter Notebook
- Git & GitHub

# 📦 Step 1: Import Libraries

In this step, we import the **NumPy** library, which provides powerful tools for numerical computations and data analysis.

```python
import numpy as np
```

# 📂 Step 2: Load the Dataset

The dataset is stored in a CSV file containing two columns:

- **Day**
- **Sales**

We use `np.genfromtxt()` to load the dataset into a NumPy array.

```python
# Load the dataset
sales_data = np.genfromtxt(
    "dataset/sales.csv",
    delimiter=",",
    skip_header=1
)
```

# 🔍 Step 3: Explore the Dataset

Before starting the analysis, let's inspect the dataset to understand its structure, dimensions, and data types.

```python
print("========== DATASET ==========")
print(sales_data)

print("\n========== DATASET INFORMATION ==========")
print(f"Type      : {type(sales_data)}")
print(f"Shape     : {sales_data.shape}")
print(f"Data Type : {sales_data.dtype}")
```

# 📈 Step 4: Extract the Sales Column

The dataset contains two columns:

- Column 0 → Day
- Column 1 → Sales

For analysis, we extract the **Sales** column into a separate NumPy array.

```python
# Extract day and sales columns
days = sales_data[:, 0]
sales = sales_data[:, 1]

print("========== SALES DATA ==========")
print(sales)
```

# 💰 Step 5: Total Revenue Analysis

## Business Problem

The company wants to know the **total revenue** generated during the month.

### NumPy Function Used

- `np.sum()`

### Expected Outcome

Calculate the total monthly revenue by adding the sales of all 30 days.

```python
# Calculate the total monthly revenue
total_revenue = np.sum(sales)

print("========== TOTAL REVENUE ==========")
print(f"Total Revenue : ₹{total_revenue:.2f}")
```

### 📈 Business Insight

The total revenue represents the overall sales generated during the month. This KPI helps management evaluate the company's overall financial performance.

# 📊 Step 6: Average Daily Sales Analysis

## Business Problem

The company wants to know the **average sales per day**.

### NumPy Function Used

- `np.mean()`

### Expected Outcome

Calculate the average daily sales to understand the company's normal daily performance.

```python
# Calculate the average daily sales
average_sales = np.mean(sales)

print("========== AVERAGE DAILY SALES ==========")
print(f"Average Daily Sales : ₹{average_sales:.2f}")
```

### 📈 Business Insight

The average daily sales act as a benchmark for evaluating daily business performance. It also helps identify unusually high or low sales days.

# 🏆 Step 7: Highest & Lowest Sales Analysis

## Business Problem

Management wants to identify the **best-performing** and **worst-performing** sales days.

### NumPy Functions Used

- `np.max()`
- `np.min()`
- `np.argmax()`
- `np.argmin()`

### Expected Outcome

Find the highest and lowest sales values along with their corresponding day numbers.

```python
# Find the highest and lowest sales
highest_sales = np.max(sales)
lowest_sales = np.min(sales)

# Find the corresponding day numbers
highest_sales_day = int(days[np.argmax(sales)])
lowest_sales_day = int(days[np.argmin(sales)])

print("========== HIGHEST & LOWEST SALES ==========")
print(f"Highest Sales : ₹{highest_sales:.2f} (Day {highest_sales_day})")
print(f"Lowest Sales  : ₹{lowest_sales:.2f} (Day {lowest_sales_day})")
```

### 📈 Business Insight

Identifying the highest and lowest sales days helps the company understand peak business periods and recognize days with poor performance. These insights can support better planning, promotions, and resource allocation.

# ⭐ Step 8: Above Average Sales Analysis

## Business Problem

The company wants to know **how many days performed better than the average daily sales**.

### NumPy Functions Used

- `np.mean()`
- Boolean Masking
- `np.sum()`

### Expected Outcome

Count the number of days where sales are greater than the average sales.

```python
# Count the number of days with above-average sales
days_above_average = np.sum(sales > average_sales)

print("========== ABOVE AVERAGE SALES ==========")
print(f"Average Sales          : ₹{average_sales:.2f}")
print(f"Days Above Average     : {days_above_average}")
```

### 📈 Business Insight

This analysis helps identify how consistently the business performs above its normal daily sales. A higher number of above-average days indicates stronger and more stable business performance.

# 📅 Step 9: Weekend vs Weekday Analysis

## Business Problem

The company wants to compare sales on **weekends** and **weekdays** to understand customer purchasing behavior.

### NumPy Concepts Used

- Boolean Masking
- Conditional Operators
- `np.sum()`

### Expected Outcome

Calculate the total sales for weekends and weekdays and determine which period performs better.

```python
# Create boolean masks
weekend_mask = (days % 7 == 6) | (days % 7 == 0)
weekday_mask = ~weekend_mask

# Extract weekend and weekday sales
weekend_sales = sales[weekend_mask]
weekday_sales = sales[weekday_mask]

# Calculate total sales
total_weekend_sales = np.sum(weekend_sales)
total_weekday_sales = np.sum(weekday_sales)

print("========== WEEKEND VS WEEKDAY ANALYSIS ==========")
print(f"Weekend Sales : ₹{total_weekend_sales:.2f}")
print(f"Weekday Sales : ₹{total_weekday_sales:.2f}")

if total_weekend_sales > total_weekday_sales:
    print("Better Performance : Weekend")
else:
    print("Better Performance : Weekday")
```

### 📈 Business Insight

Comparing weekend and weekday sales helps businesses understand customer shopping patterns. These insights can be used to schedule promotions, optimize staffing, and improve inventory planning.

# 📈 Step 10: Sales Growth Analysis

## Business Problem

The company wants to analyze how sales change from one day to the next.

### NumPy Function Used

- `np.diff()`

### Expected Outcome

Calculate the day-to-day change in sales and identify the overall sales trend.

```python
# Calculate day-to-day sales growth
daily_growth = np.diff(sales)

print("========== SALES GROWTH ANALYSIS ==========")
print("Daily Sales Growth:")
print(daily_growth)
```

### 📈 Business Insight

Daily sales growth shows whether sales are increasing or decreasing compared to the previous day. It helps management identify trends, sudden changes, and periods of strong or weak business performance.

# 📉 Step 11: Sales Performance Analysis

## Business Problem

The company wants to evaluate its day-to-day sales performance by identifying how many days sales increased, decreased, or remained unchanged.

### NumPy Functions Used

- `np.diff()`
- Boolean Masking
- `np.sum()`

### Expected Outcome

Determine the number of days with increasing, decreasing, and unchanged sales.

```python
# Calculate day-to-day sales difference
daily_growth = np.diff(sales)

# Count increase, decrease, and no-change days
increase_days = np.sum(daily_growth > 0)
decrease_days = np.sum(daily_growth < 0)
no_change_days = np.sum(daily_growth == 0)

print("========== SALES PERFORMANCE ANALYSIS ==========")
print(f"Days with Increased Sales : {increase_days}")
print(f"Days with Decreased Sales : {decrease_days}")
print(f"Days with No Change       : {no_change_days}")
```

### 📈 Business Insight

Monitoring daily sales performance helps the company understand business consistency. A higher number of increasing sales days indicates positive business growth, while frequent decreases may highlight periods requiring attention.

# 🎯 Step 12: Sales Target Analysis

## Business Problem

The company has set a daily sales target of **₹3000**. The objective is to evaluate how often this target was achieved.

### NumPy Concepts Used

- Boolean Masking
- Conditional Filtering
- `np.sum()`

### Expected Outcome

Identify the number of target achieved and target missed days along with their total sales.

```python
# Set the daily sales target
sales_target = 3000

# Count target achieved and missed days
target_achieved_days = np.sum(sales >= sales_target)
target_missed_days = np.sum(sales < sales_target)

# Calculate total sales
total_target_achieved_sales = np.sum(sales[sales >= sales_target])
total_target_missed_sales = np.sum(sales[sales < sales_target])

print("========== SALES TARGET ANALYSIS ==========")
print(f"Daily Sales Target : ₹{sales_target}")

print(f"\nDays Achieved Target : {target_achieved_days}")
print(f"Days Missed Target   : {target_missed_days}")

print(f"\nTotal Sales on Target Achieved Days : ₹{total_target_achieved_sales}")
print(f"Total Sales on Target Missed Days   : ₹{total_target_missed_sales}")
```

### 📈 Business Insight

Sales target analysis measures how consistently the business meets its expected performance. This information can help management evaluate sales effectiveness and identify areas for improvement.

# 📊 Step 13: Sales Range Analysis

## Business Problem

The company wants to identify days where sales fall within the normal operating range of **₹2000 to ₹4000**.

### NumPy Concepts Used

- Boolean Masking
- Multiple Conditions (`&`)

### Expected Outcome

List all normal sales days and count how many days fall within the specified range.

```python
# Define the normal sales range
min_sales = 2000
max_sales = 4000

# Create a boolean mask
normal_mask = (sales >= min_sales) & (sales <= max_sales)

# Extract day numbers and sales values
normal_days = days[normal_mask]
normal_sales = sales[normal_mask]

print("========== SALES RANGE ANALYSIS ==========")
print(f"Sales Range : ₹{min_sales} - ₹{max_sales}")
print(f"Number of Normal Sales Days : {len(normal_days)}")

print("\nNormal Sales Days")

for day, sale in zip(normal_days, normal_sales):
    print(f"Day {int(day)} : ₹{sale}")
```

### 📈 Business Insight

Filtering sales within a normal range helps the company understand routine business performance and identify days with unusually high or low sales.

# 🏆 Step 14: Top 5 Best Sales Days

## Business Problem

The company wants to identify its top-performing sales days.

### NumPy Functions Used

- `np.argsort()`

### Expected Outcome

Display the top five sales days in descending order.

```python
# Get indices of the top 5 sales values
top_5_indices = np.argsort(sales)[-5:][::-1]

# Extract day numbers and sales
top_5_days = days[top_5_indices]
top_5_sales = sales[top_5_indices]

print("========== TOP 5 BEST SALES DAYS ==========")

for rank, (day, sale) in enumerate(zip(top_5_days, top_5_sales), start=1):
    print(f"Rank {rank}: Day {int(day)} - ₹{sale}")
```

### 📈 Business Insight

Identifying the highest-performing sales days helps businesses understand peak demand periods, evaluate successful marketing campaigns, and improve future planning.

# 📋 Step 15: Final Business Dashboard

### Dashboard Includes

```python
# =====================================
# Final Business Dashboard
# =====================================

print("=" * 60)
print("         MONTHLY SALES BUSINESS DASHBOARD")
print("=" * 60)

print("\n📊 Dataset Summary")
print("-" * 60)
print(f"Total Days Analyzed        : {len(sales)}")

print("\n💰 Revenue Summary")
print("-" * 60)
print(f"Total Revenue              : ₹{total_revenue:.2f}")
print(f"Average Daily Sales        : ₹{average_sales:.2f}")

print("\n🏆 Sales Performance")
print("-" * 60)
print(f"Highest Sales              : ₹{highest_sales:.2f}")
print(f"Highest Sales Day          : Day {highest_sales_day}")
print(f"Lowest Sales               : ₹{lowest_sales:.2f}")
print(f"Lowest Sales Day           : Day {lowest_sales_day}")

print("\n📈 Business Insights")
print("-" * 60)
print(f"Days Above Average         : {days_above_average}")
print(f"Target Achieved Days       : {target_achieved_days}")
print(f"Target Missed Days         : {target_missed_days}")

print("\n📉 Sales Trend")
print("-" * 60)
print(f"Increase Days             : {increase_days}")
print(f"Decrease Days             : {decrease_days}")
print(f"No Change Days            : {no_change_days}")

print("\n🥇 Best Performing Day")
print("-" * 60)
print(f"Day                       : {int(top_5_days[0])}")
print(f"Sales                     : ₹{top_5_sales[0]:.2f}")

print("\n" + "=" * 60)
print("            END OF SALES REPORT")
print("=" * 60)
```
