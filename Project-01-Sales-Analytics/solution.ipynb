 Import the NumPy library for numerical computations

```python
import numpy as np
```

 Load the sales dataset from the CSV file

```python
# delimiter="," tells NumPy that values are separated by commas
# skip_header=1 ignores the first row (Day,Sales)
sales_data = np.genfromtxt(
    "dataset/sales.csv",
    delimiter=",",
    skip_header=1
)

# Display the complete dataset
print(sales_data)

# Display the data type of the object
print(type(sales_data))

# Display the dimensions of the dataset (rows, columns)
print(sales_data.shape)

# Display the data type of the array elements
print(sales_data.dtype)

# Extract only the Sales column
# ':' selects all rows
# '1' selects the second column (Sales)
sales = sales_data[:, 1]

# Display the extracted sales data
print(sales)
```

Task 1 : Total revenue analysis

```python
# Calculate the total revenue
total_revenue = np.sum(sales)
print(f"Total Revenue: ${total_revenue:.2f}")
```

Task 2 : Calculate Average analysis

```python
# Calculate the average daily sales
average_sales = np.mean(sales)
print(f"Average Daily Sales: ${average_sales:.2f}")
```

Task 3 : Highest & Lowest Sales Analysis

```python
# Highest & Lowest Sales Day
highest_sales = np.max(sales)
lowest_sales = np.min(sales)
highest_index = np.argmax(sales)
lowest_index = np.argmin(sales)
print(f"Highest Sales Day: ${highest_sales:.2f} (Day {highest_index + 1})")
print(f"Lowest Sales Day: ${lowest_sales:.2f} (Day {lowest_index + 1})")
```

Task 4 :  Above Average Sales Analysis

```python
# Days Above Average Sales
days_above_average = np.sum(sales > average_sales)
print(f"Number of Days with Sales Above Average: {days_above_average}")
```

Task 5 : Weekend vs Weekday Analysis

```python

# Weekend vs Weekday Analysis
# solution 1
# Extract Saturday sales
saturday_sales = sales[5::7]

# Extract Sunday sales
sunday_sales = sales[6::7]

# Combine both arrays
weekend_sales = np.concatenate((saturday_sales, sunday_sales))

# Total weekend sales
total_weekend_sales = np.sum(weekend_sales)

print("Weekend Sales:", total_weekend_sales)


# solution 2
# =====================================
# Weekend vs Weekday Analysis
# =====================================

# Extract day numbers
days = sales_data[:, 0]

# Create masks
weekend_mask = (days % 7 == 6) | (days % 7 == 0)
weekday_mask = ~weekend_mask

# Extract sales
weekend_sales = sales[weekend_mask]
weekday_sales = sales[weekday_mask]

# Calculate totals
total_weekend_sales = np.sum(weekend_sales)
total_weekday_sales = np.sum(weekday_sales)

# Display results
print("========== WEEKEND ANALYSIS ==========")
print(f"Weekend Sales : ₹{total_weekend_sales}")
print(f"Weekday Sales : ₹{total_weekday_sales}")

if total_weekend_sales > total_weekday_sales:
    print("Better Performance : Weekend")
else:
    print("Better Performance : Weekday")
```

Task 6 : Sales Growth Analysis

```python
# Sales Growth Analysis
# np.diff() is a NumPy function specifically designed to calculate the difference between consecutive elements. It makes the code cleaner, more readable, and avoids writing manual subtraction or loops.

# Calculate the difference in sales between consecutive days
daily_growth = np.diff(sales)

# Display the daily sales growth
print("========== SALES GROWTH ANALYSIS ==========")
print("Daily Growth:", daily_growth)

# Count the number of days with increased sales
increase_days = np.sum(daily_growth > 0)

# Count the number of days with decreased sales
decrease_days = np.sum(daily_growth < 0)

# Count the number of days with no change in sales
no_change_days = np.sum(daily_growth == 0)

# Display the results
print(f"Days with Increased Sales : {increase_days}")
print(f"Days with Decreased Sales : {decrease_days}")
print(f"Days with No Change       : {no_change_days}")
```

Task 7 : Sales Performance Analysis

```python
# Calculate the day-to-day difference in sales
daily_growth = np.diff(sales)

# Count the number of days where sales increased
increase_days = np.sum(daily_growth > 0)

# Count the number of days where sales decreased
decrease_days = np.sum(daily_growth < 0)

# Count the number of days where sales remained the same
no_change_days = np.sum(daily_growth == 0)

# Display the results
print("\n========== SALES PERFORMANCE ANALYSIS ==========")
print(f"Days with Increased Sales : {increase_days}")
print(f"Days with Decreased Sales : {decrease_days}")
print(f"Days with No Change       : {no_change_days}")

# Display the business interpretation
if increase_days > decrease_days:
    print("\nBusiness Insight:")
    print("The company experienced more days with increasing sales than decreasing sales, indicating an overall positive sales trend.")
elif decrease_days > increase_days:
    print("\nBusiness Insight:")
    print("The company experienced more days with decreasing sales than increasing sales, indicating a declining sales trend.")
else:
    print("\nBusiness Insight:")
    print("The number of increasing and decreasing sales days is equal, indicating stable sales performance.")
```

Task 8 :  Sales Target Analysis

```python
# Set the daily sales target
sales_target = 3000

# Count the number of days that achieved the target
target_achieved_days = np.sum(sales >= sales_target)

# Count the number of days that missed the target
target_missed_days = np.sum(sales < sales_target)

# Extract the sales values that achieved the target
target_achieved_sales = sales[sales >= sales_target]

# Extract the sales values that missed the target
target_missed_sales = sales[sales < sales_target]

# Calculate the total sales for achieved and missed target days
total_target_achieved_sales = np.sum(target_achieved_sales)
total_target_missed_sales = np.sum(target_missed_sales)

# Display the results
print("\n========== SALES TARGET ANALYSIS ==========")
print(f"Daily Sales Target : ₹{sales_target}")

print(f"\nDays Achieved Target : {target_achieved_days}")
print(f"Days Missed Target   : {target_missed_days}")

print("\nSales That Achieved the Target:")
print(target_achieved_sales)

print("\nSales That Missed the Target:")
print(target_missed_sales)

print(f"\nTotal Sales on Target Achieved Days : ₹{total_target_achieved_sales}")
print(f"Total Sales on Target Missed Days   : ₹{total_target_missed_sales}")
```

Task 9 :  Sales Range Analysis

```python
# Define the sales range
min_sales = 2000
max_sales = 4000

# Create a boolean mask for normal sales
normal_mask = (sales >= min_sales) & (sales <= max_sales)

# Extract the day numbers
normal_days = days[normal_mask]

# Extract the sales values
normal_sales = sales[normal_mask]

# Count the number of normal sales days
number_of_normal_days = np.sum(normal_mask)

# Display the results
print("\n========== SALES RANGE ANALYSIS ==========")
print(f"Sales Range : ₹{min_sales} - ₹{max_sales}")
print(f"Number of Normal Sales Days : {number_of_normal_days}")

print("\nNormal Sales Days and Sales:")

for day, sale in zip(normal_days, normal_sales):
    print(f"Day {int(day)} : ₹{sale}")
```

Task 10 : Top 5 Best Sales Days

```python
# Get the indices of the top 5 highest sales
top_5_indices = np.argsort(sales)[-5:][::-1]

# Extract the corresponding sales values
top_5_sales = sales[top_5_indices]

# Extract the corresponding day numbers
top_5_days = days[top_5_indices]

# Display the results
print("\n========== TOP 5 BEST SALES DAYS ==========")

for rank, (day, sale) in enumerate(zip(top_5_days, top_5_sales), start=1):
    print(f"Rank {rank}: Day {int(day)} - Sales: ₹{sale}")
```

Task 11 : Business Dashboard

```python
# Final Business Summary Dashboard

print("\n" + "=" * 50)
print("      MONTHLY SALES BUSINESS REPORT")
print("=" * 50)

print(f"Total Days Analyzed           : {len(sales)}")

print(f"Total Revenue                 : ₹{total_revenue}")

print(f"Average Daily Sales           : ₹{average_sales:.2f}")

print(f"Highest Sales                 : ₹{highest_sales}")
print(f"Highest Sales Day             : Day {highest_index}")

print(f"Lowest Sales                  : ₹{lowest_sales}")
print(f"Lowest Sales Day              : Day {lowest_index}")

print(f"Days Above Average            : {days_above_average}")

print(f"Target Achieved Days          : {target_achieved_days}")
print(f"Target Missed Days            : {target_missed_days}")

print(f"Sales Increase Days           : {increase_days}")
print(f"Sales Decrease Days           : {decrease_days}")
print(f"No Change Days                : {no_change_days}")

print("\nTop Performing Day")

print(f"Day                           : {int(top_5_days[0])}")
print(f"Sales                         : ₹{top_5_sales[0]}")

print("=" * 50)
print("End of Report")
print("=" * 50)
```
