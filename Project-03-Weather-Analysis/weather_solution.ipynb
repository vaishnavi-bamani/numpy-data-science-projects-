# 📂 Step 2: Load the Weather Dataset

The weather dataset contains daily observations collected over a period of **30 days**.

Each record includes:

- 🌡️ Temperature (°C)
- 💧 Humidity (%)
- 🌧️ Rainfall (mm)
- 💨 Wind Speed (km/h)

The dataset is stored in **CSV (Comma-Separated Values)** format.

We will use **`np.genfromtxt()`** to load the CSV file into a NumPy array. This function reads numerical data efficiently and converts it into a structured array that can be analyzed using NumPy operations.

### 📌 NumPy Function Used

- **`np.genfromtxt()`** — Reads data from a text or CSV file and converts it into a NumPy array.

### 🎯 Expected Outcome

After loading the dataset, we will have a NumPy array that can be used for statistical analysis and weather insights.

```python
weather_data = np.genfromtxt(
    "dataset/weather_analysis.csv",
    delimiter=",",
    skip_header=1
)
```

# 📖 Step 4: Extract Individual Columns

The weather dataset consists of five columns, each representing a specific weather parameter.

| Column | Description |
|---------|-------------|
| Day | Day Number |

Extracting individual columns makes the code more readable and allows us to perform statistical operations independently on each weather parameter.

# 🌡️ Step 6: Temperature Analysis

Temperature is one of the most important weather parameters, as it directly affects climate, agriculture, human health, and daily activities.

In this step, we will perform a statistical analysis of the temperature data to understand the overall weather trend during the 30-day period.

---

## 📌 Analysis to Perform

We will calculate:

- Highest Temperature
- Lowest Temperature
- Average Temperature
- Median Temperature
- Standard Deviation of Temperature

---

## 📌 NumPy Functions Used

| Function | Purpose |
|----------|---------|
| `np.max()` | Finds the highest temperature |
| `np.min()` | Finds the lowest temperature |
| `np.mean()` | Calculates the average temperature |
| `np.median()` | Finds the middle value of the temperature data |
| `np.std()` | Measures how much the temperatures vary from the average |
| `np.argmax()` | Finds the day with the highest temperature |
| `np.argmin()` | Finds the day with the lowest temperature |

---

## 🎯 Expected Outcome

After completing this analysis, we will be able to:

- Identify the hottest day.
- Identify the coldest day.
- Understand the average weather conditions.
- Measure temperature consistency using standard deviation.
- Gain valuable insights into the month's temperature distribution.

```python
# =====================================
# Temperature Analysis
# =====================================

# Calculate temperature statistics

highest_temperature = np.max(temperature)
lowest_temperature = np.min(temperature)

average_temperature = np.mean(temperature)
median_temperature = np.median(temperature)
std_temperature = np.std(temperature)

highest_temp_day = days[np.argmax(temperature)]
lowest_temp_day = days[np.argmin(temperature)]

print("========== TEMPERATURE ANALYSIS ==========\n")

print(f"Highest Temperature : {highest_temperature:.0f}°C (Day {int(highest_temp_day)})")
print(f"Lowest Temperature  : {lowest_temperature:.0f}°C (Day {int(lowest_temp_day)})")
print(f"Average Temperature : {average_temperature:.2f}°C")
print(f"Median Temperature  : {median_temperature:.2f}°C")
print(f"Standard Deviation  : {std_temperature:.2f}")
```

# 💧 Step 7: Humidity Analysis

Humidity represents the amount of moisture present in the atmosphere.

Analyzing humidity helps understand weather comfort and rainfall patterns.

### NumPy Functions Used

- np.max()
- np.min()
- np.mean()

```python
# =====================================
# Humidity Analysis
# =====================================

highest_humidity = np.max(humidity)
lowest_humidity = np.min(humidity)
average_humidity = np.mean(humidity)

print("========== HUMIDITY ANALYSIS ==========\n")

print(f"Highest Humidity : {highest_humidity:.0f}%")
print(f"Lowest Humidity  : {lowest_humidity:.0f}%")
print(f"Average Humidity : {average_humidity:.2f}%")
```

# 🌧️ Step 8: Rainfall Analysis using np.where()

Rainfall is classified into different categories using conditional selection.

### Categories

- Dry Day
- Rainy Day
- Heavy Rainfall Day

### NumPy Function Used

- np.where()

# =====================================
# Rainfall Analysis
# =====================================

dry_days = np.where(rainfall == 0)[0]
rainy_days = np.where((rainfall > 0) & (rainfall <= 10))[0]
heavy_rain_days = np.where(rainfall > 10)[0]

print("========== RAINFALL ANALYSIS ==========\n")

print(f"Dry Days            : {len(dry_days)}")
print(f"Rainy Days          : {len(rainy_days)}")
print(f"Heavy Rainfall Days : {len(heavy_rain_days)}")

# 💨 Step 9: Wind Speed Analysis

Wind speed is classified into:

- Calm
- Moderate
- Windy

using boolean masking.

```python
# =====================================
# Wind Speed Analysis
# =====================================

calm_days = np.sum(wind_speed < 15)
moderate_days = np.sum((wind_speed >= 15) & (wind_speed <= 20))
windy_days = np.sum(wind_speed > 20)

print("========== WIND SPEED ANALYSIS ==========\n")

print(f"Calm Days      : {calm_days}")
print(f"Moderate Days  : {moderate_days}")
print(f"Windy Days     : {windy_days}")
```

# 🌤️ Step 10: Temperature Classification

Using **np.where()**, classify each day as:

- Cold
- Pleasant
- Hot

```python
# =====================================
# Temperature Classification
# =====================================

temperature_status = np.where(
    temperature < 28,
    "Cold",
    np.where(temperature <= 33, "Pleasant", "Hot")
)

print("========== TEMPERATURE CLASSIFICATION ==========\n")

for day, temp, status in zip(days, temperature, temperature_status):
    print(f"Day {int(day):2d} : {temp:.0f}°C -> {status}")
```

# 📊 Step 11: Weather Statistics

Calculate statistical values that describe the distribution of temperatures.

### NumPy Functions Used

- np.percentile()
- np.sort()

```python
# =====================================
# Weather Statistics
# =====================================

sorted_temperature = np.sort(temperature)

q1 = np.percentile(temperature,25)
median = np.percentile(temperature,50)
q3 = np.percentile(temperature,75)

print("========== WEATHER STATISTICS ==========\n")

print("Sorted Temperatures")
print(sorted_temperature)

print()

print(f"25th Percentile : {q1:.2f}")
print(f"Median          : {median:.2f}")
print(f"75th Percentile : {q3:.2f}")
```

# 📋 Step 12: Final Weather Dashboard

The dashboard summarizes all important weather statistics obtained from the dataset.

```python
# =====================================
# Final Weather Dashboard
# =====================================

print("="*60)
print("          WEATHER ANALYSIS REPORT")
print("="*60)

print(f"Total Days                  : {len(days)}")

print(f"Highest Temperature         : {highest_temperature:.0f}°C")
print(f"Lowest Temperature          : {lowest_temperature:.0f}°C")

print(f"Average Temperature         : {average_temperature:.2f}°C")
print(f"Median Temperature          : {median_temperature:.2f}°C")

print(f"Standard Deviation          : {std_temperature:.2f}")

print()

print(f"Highest Humidity            : {highest_humidity:.0f}%")
print(f"Lowest Humidity             : {lowest_humidity:.0f}%")
print(f"Average Humidity            : {average_humidity:.2f}%")

print()

print(f"Dry Days                    : {len(dry_days)}")
print(f"Rainy Days                  : {len(rainy_days)}")
print(f"Heavy Rainfall Days         : {len(heavy_rain_days)}")

print()

print(f"Calm Days                   : {calm_days}")
print(f"Moderate Wind Days          : {moderate_days}")
print(f"Windy Days                  : {windy_days}")

print("="*60)
print("End of Report")
print("="*60)
```
